```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('/kaggle/input/heart-diseases/heart_disease_data.csv')
df.head()
```

```python
print(df.shape)
df.describe()
```

```python
df.isnull().sum()
```

```python
y=df.target
x=df.drop(columns='target',axis=1)
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=2,test_size=0.3)
print(x.shape,xtrain.shape,xtest.shape)
```

```python
model=LogisticRegression()
model.fit(xtrain,ytrain)
```

```python
pred=model.predict(xtest)
print(accuracy_score(pred,ytest))
```

```python
t=(63,1,3,145,233,1,0,150,0,2.3,0,0,1)
nparr=np.asarray(t)
nparr=nparr.reshape(1,-1)
model.predict(nparr)
```

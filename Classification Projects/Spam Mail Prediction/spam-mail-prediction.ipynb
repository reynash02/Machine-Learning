```python
#importing modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer 
from sklearn.naive_bayes import GaussianNB
```

```python
#importing dataset
df=pd.read_csv('/kaggle/input/mail-dataset/mail_data.csv')
df.head()
```

```python
#statistical analysis
print(df.shape)
df.describe()
```

```python
df.isnull().sum()
```

```python
#encoding mail column
df.replace({'Category':{'ham':0,'spam':1}},inplace=True)
df.head()
```

```python
#splitting into training and testing data
x=df.Message
y=df.Category
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=2,test_size=0.2)
print(x.shape,xtrain.shape,xtest.shape)
print(y.shape,ytrain.shape,ytest.shape)
```

```python
#feature extraction
cv=CountVectorizer()
xtrain_count=cv.fit_transform(xtrain.values)
xtrain_count.toarray()[:3]
```

```python
#using machine learning
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(xtrain_count,ytrain)
```

```python
mails=['Hi how are you','Get 20% discount']
mailscount=cv.transform(mails)
model.predict(mailscount)
```

```python
#using testing data on the model
xtest_count=cv.transform(xtest)
model.score(xtest_count,ytest)
```

```python
#predictive model
l=[]
string=str(input('Enter your message='))
l.append(string)
str_count=cv.transform(l)
a=model.predict(str_count)
if a[0]==0:
    print('Not Spam')
elif a[0]==1:
    print('Spam')
```
